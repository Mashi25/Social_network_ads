#Preprocessing of Data

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset= pd.read_csv('/content/Social_Network_Ads.csv') 
```

```
dataset
```

```
X = dataset.iloc[:,:-1].values
y= dataset.iloc[:,-1].values
```

```
X
```

```
y
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test, y_train , y_test = train_test_split(X, y, test_size=0.25)
```

```
len(X_train)
```

```
from sklearn.preprocessing import StandardScaler
```

```
sc = StandardScaler()
```

```
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
X_test
```

#Logic Regression 

```
from sklearn.linear_model import LogisticRegression
```

```
lcla= LogisticRegression()
```

```
lcla.fit(X_train, y_train)
```

as it is stdised hence the two festures also shoould be stdised

```
lcla.predict(sc.transform([[19,19000]]))
```

```
pred=lcla.predict(X_test)
```

```
for i in range(len(pred)):
  print(pred[i] , y_test[i])
```

Confusion Matrix
Screenshot_20200901_154339.jpg

```
from sklearn.metrics import accuracy_score , confusion_matrix
```

```
ac = accuracy_score(y_test, pred)
cm = confusion_matrix(y_test, pred)
```

```
ac
```

```
cm
```

```
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, lcla.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
